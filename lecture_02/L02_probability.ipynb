Statistics is the mathematical framework used for analyzing, intepreting and presenting data. In practice, data is collected through experiments and these experiments are often subject to uncertainty, i.e. their outcome is not known prior to measurement. Probability theory is the mathematical framework in which we model uncertainty.

**How can we model something that is uncertain?** Events that are uncertain at the micro-level often appear to be deterministic at the macro-level. I.e. even though we cannot predict a single event we use experiments to draw conclusions about the underlying processes governing their outcome. This is what probability theory does.

## Notation and definitions

**Set:** A set is a collection of elements, and we will denote it by $A = \{...\}$. An example is $A = \{1, 2, 3\}$. The empty set is the unique set that contains no elements: $\varnothing = \{\}$.

**Operations on sets:**

<u>Union</u>: The union of two sets $A$ and $B$ is the set that contains all elements from $A$ and all elements from $B$, and is denoted by $A \cup B$. As an example, for $A = \{1, 3, 5\}$ and $B = \{2, 3, 4\}$, we have $A \cup B = \{1, 2, 3, 4, 5\}$.

<u>Intersection</u>: The intersection of two sets $A$ and $B$ is the set that contains all elements that are both in $A$ and $B$, and it is denoted by $A \cap B$. For our two sets from before, we have $A \cap B = \{3\}$. Two sets are said to be *disjoint* if their intersection is the empty set: $A \cap B = \varnothing$.

<u>Complement</u>: The complement of a set $A$ is the set of all elements not in $A$, and is denoted by $A^c$. This is always relative to a larger set $B$. If $B$ is just another set, then $A^c$ contains all elements in $B$ that are not in $A$. If $B$ is the set of all sets in the Universe, then $A^c$ contains all set elements that are not in $A$. As an example, $A = \{1, 3, 5\}$ and $B = \{1, 2, 3, 4, 5, 6\}$, then $A^c = \{2, 4, 6\}$. <font color='pink'>[AN: maybe put image]</font>

**Outcome spaces and events:**

Experiments are actions that can have a set of outcomes but we cannot predict the exact outcome prior to the experiment. The set of all possible outcomes for a given experiemnt is aclled the *outcome space*, which we denote by $\Omega$. For example in case of an experiment of throwing a fair dice the outcome space is $\Omega = \{1, 2, 3, 4, 5, 6\}$. An event is a subset of the outcome space. In our fair dice example, a possible event would be that the dice shows an even number, i.e. $E = \{2, 4, 6\}$. The collection of all events is called event space $\Sigma$. Note that $\Omega$ and $\Sigma$ are not the same. The outcome space contains the basic outcomes of an experiment while the event space contains the set of outcomes.

Proability theory is concerned with assigning probabilities (i.e. a scalar function) to events and of combining these probabilities to obtain the probability of complicated events. In this sense, events are subsets of the outcome space to which we can assign a probability. 

## Probability

A probability space consists of the collection $(\Omega, \mathcal{E}, P)$, where $\Omega$ is the outcome space, $\mathcal{E}$ denotes a class of events and $P$ is a probability measure. $P$ assigns a probability to all events in $\mathcal{E}$ and obeys the three Ko0lmogorov axioms of probability:

**Axiom 1:** $0 \leq P(A) \leq 1$ for all events $A \in \mathcal{E}$.

**Axiom 2:** $P(\Omega) = 1$.

**Axiom 3:** For pairwise disjoint (mutually exclusive) events $A_{1}, A_{2}, ...$, $A_{i} \cap A_{j} = \varnothing$ for $i \neq j$ we have $$P(A_{1} \cup A_{2} \cup A_{3} ...) = P(A_{1}) + P(A_{2}) + P(A_{3}) + ...,$$ or more formally $$P\left(\bigcup_{i=1}^{\infty} A_{i}\right) = \sum_{i=1}^{\infty} P(A_{i}).$$

We obtain the following corrollaries to these axioms:

For any event $A$, the probability of its complement $A^{c} = \{w \in \Omega | w \notin A\}$ satisifes: $$P(A^{c}) = 1 - P(A).$$

For any two events $A, B$, we have that $$P(A \cup B) = P(A) + P(B) - P(A\cap B).$$

### Examples

* A fair dice:

    The outcome space is $\Omega = \{1, 2, 3, 4, 5, 6 \}$, and the associated probability measure is $$P(\{i\}) = 1/6, i \in \Omega.$$


* Loaded dice:

    Similar to above, $\Omega = \{1, 2, 3, 4, 5, 6 \}$ and $$P(\{i\}) = \frac{w_i}{w_1 + w_2 + ... + w_6},$$ where $w_{i}$ are weights.
    
### Conditional probabilities

Let's start with an example using our favorite dice. Let us consider the event $A = \{1, 2, 3\}$, which has probability $P(A) = \frac{1}{2}$. Let's now assume that someone tells you that the dice actually returned an even number, i.e. $B = \{2, 4, 6\}$. How does this information now influence the assignment of probability to $A$? After knowing that $B$ happened, we have that $P(b) = \frac{1}{3}$ for each element $b \in B$ (i.e. $P(B) = 1$ and $P(B^c)=0$). Therefore, the probability of $A$ knowing that $B$ happened is given by the number of events in $A$ that are also in $B$ divided by the number of events in $B$, i.e. $P(A|B) = \frac{1}{3}$.

More generally, let us consider an experiment with $m$ outcomes that all have the same probability, and $A, B$ are two events. If we know that $B$ happened, what is the probability of $A$? We assume that $A$ contains $k$ outcomes, while $B$ contains $n$, and we have that $A \cup B$ contains $i$ elements. Therefore the probability of $A$ given $B$ can be computed as $$P(A|B) = \frac{P(A\cup B)}{P(B)},$$ which is the definition of *conditional probability* for $P(B) \neq 0$.

As a corrollary we obtain the *multiplicative rule for probability* as $$P(A\cup B) = P(A|B)P(B).$$

Two events $A, B$ are said to be independent if $$P(A\cup B) = P(A)P(B),$$ i.e. knowing $B$ does not change anything about knowing $A$, $P(A|B) = P(A)$.

### The law of total probability

As a final corrollary, we can derive the *law of total probability*. Before, we need to define a *partition*: A partition of the outcome space $\Omega$, $B_{1}, ..., B_{n}$, is a collection of pairwise disjoint sets whose union is $\Omega$, i.e. $B_{i} \cap B_{j} = \varnothing$ for $i \neq j$. 

The law of total probability now states that for any event $A \in \Omega$ and partition $B_{1}, ..., B_{n}$ we can compute $P(A)$ by combining the probabilities of $A$ for each $B_{i}$, i.e. $$P(A) = P(A \cap B_{1}) + ... + P(A \cap B_{n}).$$ Using conditional probabilities, we can rewrite this as $$P(A) = P(A|B_{1})P(B_1) + ... + P(A|B_{n})P(B_n).$$

## Bayes' Theorem

Combining the law of total probability with the definition of conditional probabilty we can derive Bayes' theorem. Let $B_{1}, ..., B_{n}$ be a partition of the outcome space. The for $i=1,...,n$ we have $$P(B_i|A)=\frac{P(A|B_{i})P(B_i)}{P(A|B_{1})P(B_1) + ... + P(A|B_{n})P(B_n)}.$$ So in essence, Bayes' theorem allows us to obtain $P(B_i|A)$ in terms of $P(A|B_{i})$ and $P(B_i)$. <font color='pink'>[AN: example]</font>

### Example: Covid-19 tests

What is the probability to have Covid-19 given you have tested positive?

Let us denote the following cases:

$T=1$: you test positive,

$T=0$: you test negative,

$D=1$: you have Covid-19,

$D=1$: you do not have Covid-19.

We are interested in the probability $P(D=1|T=1)$. Using Bayes' Theorem we have $$P(D=1|T=1)=\frac{P(T=1|D=1)P(D=1)}{P(T=1)}=\frac{P(T=1|D=1)P(D=1)}{P(T=1|D=1)P(D=1)+P(T=1|D=0)P(D=0))}.$$ Now let's look at the data:

$P(D=1)=0.01$, i.e. we assume that the chances of having Covid-19 at the time are 1 in 100

Furthermore, the test suppliers tell us the following:

$P(T=0|D=1)=0.001$: false negative rate, 

$P(T=1|D=0)=0.02$: false positive rate.

Thus we have $P(T=1|D=1)=0.999$, P(D=0)=0.99, and thus we get $$P(D=1|T=1)=\frac{0.999\times0.01}{0.999\times0.01+0.02\times0.99}=0.33.$$  


### Example: O. J. Simpson trial

The O. J. Simpson murder case was a famous criminal trial in which former American football star O.J. Simpson was tried on two counts of murder for the June 12, 1994, deaths of his
ex-wife Nicole Brown Simpson and Mezzaluna restaurant waiter Ronald Goldman. During the trial, the prosecution brought up evidence that O.J. Simpson had a history of violence toward his ex-wife. The defense argued that past domestic abuse is irrelevant because only $0.1\%$ of the men who physically abuse their wives actually end up murdering them.

Is this argument correct?

Let us use Bayes' Theorem to understand this.

We have the following events:

$A$: woman abused by her husband, boyfriend, or lover, 

$M$: woman murdered, 

$G$: woman murdered by her abuser.

We are interested in the probability of $G$, given $A$ and $M$, i.e. $P(G|M, A)$. We know that $$P(G|M, A) = \frac{P(M, A, G)}{P(A, M)}=\frac{P(M|A, G)P(G|A)P(A)}{P(M|A)P(A)}=\frac{P(M|A, G)P(G|A)}{P(M|A)}=\frac{P(M|A, G)P(G|A)}{P(M|A, G)P(G|A)+P(M|A, G^c)P(G^c|A)}.$$

Now let's look at the data. We know from the defense that $$P(G|A)=0.001,$$ and thus $$P(G^c|A)=0.999.$$ In addition, we have that $$P(M|A, G)=1.$$ What we now need is $P(M|A, G^c).$ In 1994, around 5000 women were murdered in the US, 1500 by their husband. Assuming a population of 100 million wifes in the US, we have $$P(M|A, G^c)=P(M|G^c)=\frac{3500}{100\times10^6}\simeq\frac{1}{30'000},$$ where the first equality comes from the fact that information that the woman was abused should not influence murder rates if the abuser was not the murderer.

So now putting everything together, we have $$P(G|M, A)=\frac{P(M|A, G)P(G|A)}{P(M|A, G)P(G|A)+P(M|A, G^c)P(G^c|A)}=\frac{1/1000}{1/1000+1/30'000\times999/1000}\simeq0.97.$$

### Random variable:
We define a *random variable* $X$ as a function from outcome space to a measurable space $X: \Omega \rightarrow M$. We can think of this function as assigning a real number to each outcome of an experiment, which then makes it easy to define functions over outcome space, as we can just define them over the corresponding random variable. 

So far we have been considering experiments with discrete outcomes (i.e. a countable outcome space), e.g. tossing a coin or rolling a dice. However, many experiments (e.g. weighing yourself on the scale in the morning) have an uncountable outcome space, meaning that their result can take on any rational number for example. We can define random variables both for countable and uncountable outcome spaces, which leads to *discrete* and *continuous* random variables respectively.

### Cumulative distribution function
Once we have random variables we can define functions on them. The *cumulative distribution function* (cdf) $F$ of a random variable $X$ is defined as $$F(x) = P(X\leq x) = P(\omega \in \Omega|X(w)\leq x),$$ where $x$ is a real number. It gives the probability that $X$ takes on any value smaller or equal than $x$. For a discrete random variable $X$ with outcomes $x_1, x_2, ...$ we have $$F(x) = P(X\leq x) = \sum_{x_i\leq x}P(X = x_i).$$

The cdf is a nondecreasing function and satifies: $$\lim_{x \to -\infty} F(x) = 0 \; \mathrm{and} \lim_{x \to \infty} F(x) = 1.$$

### Probability density function
While the probability for a discrete random variable to take on a specific value is nonzero, this is not the case for a continuous random variable, as there are uncountably many outcomes in any interval and thus the total probability would not sum up to 1. We can still deal with continuous random variables by defining the *probability density function* (pdf) as $$f(x) = \frac{\mathrm{d}F}{\mathrm{d}x}.$$ With this, we can express the probability for a random variable to take on a value in an interval $\mathrm{d}x$ as $$P(X \in [x, x+\mathrm{d}x])=F(x+\mathrm{d}x)-F(x)=f(x)\mathrm{d}x.$$ Thus we can also write: $$F(x) = \int_{-\infty}^{x}f(y)\mathrm{d}y.$$

We can generalize these definitions to more than one random variable, a so-called random vector. For example for a random vector $(X_1, ..., X_n)$ the probability that $X_1 \leq x_1, ..., X_n \leq x_n$ is given by $$F(x_1, ..., x_n) = P(X_1 \leq x_1, ..., X_n \leq x_n) = P(\omega \in \Omega|X_1(\omega)\leq x_1, ..., X_n(\omega)\leq x_n).$$ Analogously, the probability for $(X_1, ..., X_n)$ to lie within a volume $\mathrm{d}x_1 ... \mathrm{d}x_n$ is $f(x_1, ..., x_n)\mathrm{d}x_1...\mathrm{d}x_n.$

**Marginalization:** Let us focus on two random variables $X_1, X_2$ for simplicity. These have a joint pdf $f(x_1, x_2)$, which is a 2D function. If we are now interested in only one of these variables, i.e. its 1D distribution, we can compute the so-called *marginal diastribution* as : $$f(x_1) = \int \mathrm{d}x_2 f(x_1, x_2).$$ 



**Independence of random variables:** As before with discrete events, we define random variables $X_1, ..., X_n$ to be independent if $$P(X_1\leq x_1 ... X_n\leq x_n) = P(X_1\leq x_1) ... P(X_n\leq x_n)$$ for all real numbers $x_1, ..., x_n$. <font color='purple'>[NTS: Compare to independence of events, here needs to be for all xi, i.e. you can let it go to infty and then get the independence of all subsets also.]</font> 

## Properties of random variables

We can use the pdf $f(x)$ to learn more about the propoerties of a random variable $X$. For example, an important indicator of the central location of a random variable (the center of the pdf) is the *mean* or *expectation value*, defined as (here explicitly for a continuous 1D random variable): $$\mu = \langle X \rangle = \int \mathrm{d}x \; x f(x),$$ where we require $\int_{-\infty}^\infty \mathrm{d}x \; |x|f(x) < \infty$. Analogously, we can compute the expected value of any function $h$ of $X$, which is a random variable itself through: $$\langle h \rangle = \int \mathrm{d}x \; h(x) f(x).$$

The second moment of a distribution quantifies its spread, and is called the *variance* $\sigma^2$ and is defined as: $$\sigma^2 = \langle (X-\mu)^2 \rangle = \langle X^2 \rangle - \mu^2.$$ The standard deviation $\sigma$ is defined as the square root of the variance, i.e. $$\sigma=\sqrt{\langle (X-\mu)^2 \rangle}.$$

We can generalizing these concepts to more than one random variable. Here we look at the example of a bivariate distribution, but this can straightforwardly be generalized to more than two random variables. Let $X, Y$ be two random variables with joint pdf $f(x, y)$. The means of the two random variables can be obtained though $$\mu_x = \langle X \rangle = \int \mathrm{d}y\; \mathrm{d}x \; x f(x, y),$$ $$\mu_y = \langle Y \rangle = \int \mathrm{d}x \;\mathrm{d}y \; y f(x, y).$$ Their respective variances are then $$\sigma_x^2 = \langle (X-\mu_x)^2 \rangle = \int \mathrm{d}y\; \mathrm{d}x \; (x-\mu_x)^2 f(x, y),$$ $$\sigma_y^2 = \langle (X-\mu_y)^2 \rangle = \int \mathrm{d}x \;\mathrm{d}y \; (y-\mu_y)^2 f(x, y).$$

Finally, we can define the covariance of the two random variables $X, Y$ as: $$\mathrm{Cov}(X, Y) = \langle (X-\mu_x)(Y-\mu_y) \rangle = \int \mathrm{d}y\; \mathrm{d}x \; (x-\mu_x)(y-\mu_y) f(x, y).$$ The covariance measures the relation between the scatter in two random variables. If $\mathrm{Cov}(X, Y)=0$, the two variables are said to be uncorrelated, while $\mathrm{Cov}(X, Y) \neq 0$ means that the variations in the two variables are related. <font color='pink'>[AN: maybe do the variance of sum of rvs as example]</font>

## Example: the Gaussian distribution

### The univariate Gaussian distribution

Let $X$ be a one-dimensional Gaussian random variable. Then the probability for $X$ to take a value in $[x, x+\mathrm{d}x]$ is given by $$P(X \in [x, x+\mathrm{d}x])=f(x)\mathrm{d}x=\mathcal{N}(x)\mathrm{d}x = \frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{(x-\mu)^2}{2\sigma^2}}.$$ In the above equation, $\mu$ is the mean of the Gaussian distribution (and thus the random variable $X$), and $\sigma$ is its standard deviation. Equivalently we can also specify the variance $\sigma^2$. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gaussian(x, mu, sigma):
    
    gauss = 1./np.sqrt(2*np.pi)/sigma*np.exp(-(x-mu)**2/2./sigma**2)
    
    return gauss

mu = 0.
sigma = 0.1

x = np.linspace(-0.5, 0.5, 100)

plt.plot(x, gaussian(x, mu, sigma), color='#e3a19c')
plt.axvline(x=0, linestyle='--', color='k', alpha=0.5)
plt.axvline(x=sigma, linestyle=':', color='k', alpha=0.5)
plt.axvline(x=-sigma, linestyle=':', color='k', alpha=0.5)
plt.xlabel(r'$x$')
plt.ylabel(r'$\mathcal{N}(x)$')
```

### The multivariate Gaussian distribution

Let us now assume that we have a vector of $n$ random variables $\boldsymbol{X}=(X_1, ..., X_n)$ that follow a multivariate Gaussian distribution. Then their pdf is given by $$f(\boldsymbol{x})=\mathcal{N}(\boldsymbol{x}) = \frac{1}{\sqrt{(2\pi)^n|C|}}e^{-\frac{1}{2}(\boldsymbol{x}-\boldsymbol{\mu})^TC^{-1}(\boldsymbol{x}-\boldsymbol{\mu})},$$ where $\boldsymbol{x}$ is the vector of random variables and $\boldsymbol{\mu}$ is the vector of their means. The quantity $C$ denotes the covariance matrix, which is defined as $$C = \begin{pmatrix}
\langle (X_1-\mu_1)(X_1-\mu_1) \rangle & \langle (X_1-\mu_1)(X_2-\mu_2) \rangle & \cdots & \langle (X_1-\mu_1)(X_n-\mu_n) \rangle \\
\vdots  & \vdots  & \ddots & \vdots  \\
\langle (X_n-\mu_n)(X_1-\mu_1) \rangle & \cdots & \langle (X_n-\mu_n)(X_n-\mu_n) \rangle 
\end{pmatrix}.$$ The diagonal elements give the variances $\sigma^2_1, ..., \sigma^2_n$ of the $X_1, ..., X_n$, while the off-diagonal elements give the cross-covariances, i.e. a measure of the correlation between the different random variables. 

In the bivariate case, this simplifies to $$C = \begin{pmatrix}
\langle (X_1-\mu_1)(X_1-\mu_1) \rangle & \langle (X_1-\mu_1)(X_2-\mu_2) \rangle \\
\langle (X_n-\mu_2)(X_1-\mu_1) \rangle & \langle (X_2-\mu_2)(X_2-\mu_2) \rangle
\end{pmatrix}.$$ We can define the correlation coefficient of two random variables $X_1, X_2$ as $$r=\frac{\langle (X_1-\mu_1)(X_2-\mu_2) \rangle}{\sigma_1\sigma_2},$$ which allows us to write the covariance matrix as $$C = \begin{pmatrix}
\sigma^2_1 & r\sigma_1\sigma_2  \\
r\sigma_1\sigma_2 & \sigma^2_2 
\end{pmatrix}.$$ The correlation coefficient quantifies the correlation between two random variables. It satisifes $r\in [-1, 1]$, where $r=0$ denotes uncorrelated variables while $r=1, r=-1$ denote full correlation or anti-correlation respectively.

Note that even though we have introduced the covariance matrix in the context of Gaussian data, we can define the covariance matrix for any vector of random variables and view it as the generalization of the variance.
